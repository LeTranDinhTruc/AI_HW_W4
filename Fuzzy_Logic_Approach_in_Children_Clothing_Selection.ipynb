```python
from numpy.core.multiarray import dtype
# Medical Diagnosis System Using Fuzzy Logic
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

height=ctrl.Antecedent(np.arange(93.5,164,0.01),'height')
chest=ctrl.Antecedent(np.arange(49.3,82.5,0.01),'chest')
waist=ctrl.Antecedent(np.arange(46.75,67.25,0.01),'waist')
size=ctrl.Consequent(np.arange(4,14.1,0.1),'size')


height['4-5']=fuzz.trimf(height.universe,[93.5,106.15,110])
height['5-6']=fuzz.trimf(height.universe,[93.5,111.94,116])
height['6-7']=fuzz.trimf(height.universe,[93.5,117.73,122])
height['7-8']=fuzz.trimf(height.universe,[93.5,123.52,128])
height['8-9']=fuzz.trimf(height.universe,[93.5,129.31,134])
height['9-10']=fuzz.trimf(height.universe,[93.5,135.1,140])
height['10-11']=fuzz.trimf(height.universe,[93.5,140.89,146])
height['11-12']=fuzz.trimf(height.universe,[93.5,146.68,152])
height['12-13']=fuzz.trimf(height.universe,[93.5,152.47,158])
height['13-14']=fuzz.trimf(height.universe,[93.5,158.26,164])

chest['4-5']=fuzz.trimf(chest.universe,[49.3,55.97,58])
chest['5-6']=fuzz.trimf(chest.universe,[49.3,58.38,60.5])
chest['6-7']=fuzz.trimf(chest.universe,[49.3,60.8,63])
chest['7-8']=fuzz.trimf(chest.universe,[49.3,62.73,65])
chest['8-9']=fuzz.trimf(chest.universe,[49.3,66.1,68.5])
chest['9-10']=fuzz.trimf(chest.universe,[49.3,69,71.5])
chest['10-11']=fuzz.trimf(chest.universe,[49.3,71.89,74.5])
chest['11-12']=fuzz.trimf(chest.universe,[49.3,75.27,78])
chest['12-13']=fuzz.trimf(chest.universe,[49.3,77.2,80])
chest['13-14']=fuzz.trimf(chest.universe,[49.3,79.61,82.5])


waist['4-5']=fuzz.trimf(waist.universe,[46.75,53.08,55])
waist['5-6']=fuzz.trimf(waist.universe,[46.75,54.04,56])
waist['6-7']=fuzz.trimf(waist.universe,[46.75,55.01,57])
waist['7-8']=fuzz.trimf(waist.universe,[46.75,55.97,58])
waist['8-9']=fuzz.trimf(waist.universe,[46.75,57.9,60])
waist['9-10']=fuzz.trimf(waist.universe,[46.75,59.59,61.75])
waist['10-11']=fuzz.trimf(waist.universe,[46.75,61.76,64])
waist['11-12']=fuzz.trimf(waist.universe,[46.75,63.21,65.5])
waist['12-13']=fuzz.trimf(waist.universe,[46.75,64.17,66.5])
waist['13-14']=fuzz.trimf(waist.universe,[46.75,64.9,67.25])

size['4-5']=fuzz.trapmf(size.universe,[4,4,5,6])
size['5-6']=fuzz.trapmf(size.universe,[4,5,6,8])
size['6-7']=fuzz.trapmf(size.universe,[4,6,7,9])
size['7-8']=fuzz.trapmf(size.universe,[4,7,8,10])
size['8-9']=fuzz.trapmf(size.universe,[4,8,9,11])
size['9-10']=fuzz.trapmf(size.universe,[4,9,10,12])
size['10-11']=fuzz.trapmf(size.universe,[4,10,11,13])
size['11-12']=fuzz.trapmf(size.universe,[4,11,12,14])
size['12-13']=fuzz.trapmf(size.universe,[4,12,13,14])
size['13-14']=fuzz.trapmf(size.universe,[4,13,14,14])

chest.view()
waist.view()
height.view()
size.view()


rule_1=ctrl.Rule(height['4-5']&chest['4-5']&waist['4-5'],size['4-5'])
rule_2=ctrl.Rule(height['5-6']&chest['5-6']&waist['5-6'],size['5-6'])
rule_3=ctrl.Rule(height['6-7']&chest['6-7']&waist['6-7'],size['6-7'])
rule_4=ctrl.Rule(height['7-8']&chest['7-8']&waist['7-8'],size['7-8'])
rule_5=ctrl.Rule(height['8-9']&chest['8-9']&waist['8-9'],size['8-9'])
rule_6=ctrl.Rule(height['9-10']&chest['9-10']&waist['9-10'],size['9-10'])
rule_7=ctrl.Rule(height['10-11']&chest['10-11']&waist['10-11'],size['10-11'])
rule_8=ctrl.Rule(height['11-12']&chest['11-12']&waist['11-12'],size['11-12'])
rule_9=ctrl.Rule(height['12-13']&chest['12-13']&waist['12-13'],size['12-13'])
rule_10=ctrl.Rule(height['13-14']&chest['13-14']&waist['13-14'],size['13-14'])

size_control=ctrl.ControlSystem([rule_1,rule_2,rule_3,rule_4,rule_5,rule_6,rule_7,rule_8,rule_9,rule_10])
size_c=ctrl.ControlSystemSimulation(size_control)
size_c.input['height']=135
size_c.input['chest']=60
size_c.input['waist']=59
size_c.compute()
print(size_c.output['size'])
size.view(sim=size_c)
```
